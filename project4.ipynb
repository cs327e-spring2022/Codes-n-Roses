##### 2) create and populate BQ tables:

```python
import os
```

```python
gsutil_cmd = "gsutil ls gs://cs327e-open-access/school-enrollments/*"
file_listings = os.popen(gsutil_cmd)

for file in file_listings:

    start_index = file.rindex("_") + 1
    end_index = file.rindex(".")
    table = file[start_index:end_index]
    
    if table in ("co", "district", "fl", "ga", "id", "ma", "mo", "ne", "ny", "or", "school", "sd", "wi"):
        print("skipping " + table)
        continue

    bq_cmd = "bq --location=US load --autodetect --skip_leading_rows=1 "\
             "--source_format=CSV " + dataset_id + "." + table + ' ' + file
    print(bq_cmd)
    
    os.system(bq_cmd)
```

#### Note: STRING type for "total" column

```python
%%bigquery
select school, grade, year, sum(cast(replace(replace(total, ",", ""), "<", "") as int64)) as total
from `universal-torch-339209.school_enrollments.tx`
where total != "" and school != "" and district like '%AUSTIN%'
group by school, grade, year
order by school, grade, year
```

This query reports the total number of hispanic students in school districts in Nevada in the year 2019 from grades pre-kindergarten to 12th grade.

Wyoming and oklahoma school districts with more than 100 asian students in schools (found only one :( )

```python
%%bigquery 
select district, AVG(asian) from school_enrollments.wy group by district having AVG(asian) > 100
UNION ALL 
select district, AVG(asian) from school_enrollments.ok group by district having AVG(asian)> 100
order by district;
```

Wyoming and oklahoma school districts with more than 100 white students in schools

```python
%%bigquery 
select district, AVG(white) from school_enrollments.wy group by district having AVG(white) > 100
UNION ALL 
select district, AVG(white) from school_enrollments.ok group by district having AVG(white)> 100
order by district;
```

Nevada school districts with more than a total of 100 black students after 2019

```python
%%bigquery 
select school, sum(black) from school_enrollments.nv 
where grade = 'k_12_total' and year >= 2019
group by school having sum(black) > 100 
order by sum(black);
```

Nevada school districts with more than a total of 300 hispanic students after 2019 

```python
%%bigquery 
select school, sum(hispanic) from school_enrollments.nv 
where grade = 'k_12_total' and year >= 2019
group by school having sum(hispanic) > 300 
order by sum(hispanic);
```
