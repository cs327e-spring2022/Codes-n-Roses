```python
## INNER JOIN SQL COMMANDS
```

```python
## Using INNER JOIN to get Movies that have a PG rating and were English movies (language_id) and ordered from least to greatest by length of the movies.
```

```python
!psql -c "select film.film_id, film.title, film.rating, film.length, film.special_features, film.language_id, language.name \
            FROM pagila.film INNER JOIN pagila.language ON film.language_id = language.language_id \
            WHERE film.rating = 'PG' ORDER BY film.length LIMIT 10"
```

```python
## Using INNER JOIN to get employees that have a store_id less than 3 and also get their firstname, lastname, email, username, and password from FK (store_id) between tables store and staff and ordered alphabetically by last name.
```

```python

!psql -c "select store.store_id, store.last_update, staff.first_name, staff.last_name, staff.email, staff.username, staff.password \
            FROM pagila.store INNER JOIN pagila.staff ON store.store_id = staff.store_id \
            WHERE store.store_id < 3 ORDER BY staff.last_name LIMIT 2"
```

```python
## Using INNER JOIN to get rentals that have a retnal_id less than 6 and also get their return_date, rental_date and last_update from FK (inventory_id) between tables rental and inventory and ordered by return date of the rentals.
```

```python
!psql -c "select rental.rental_id, rental.return_date, rental.rental_date, rental.last_update\
            FROM pagila.rental INNER JOIN pagila.inventory ON rental.inventory_id = inventory.inventory_id \
            WHERE rental.rental_id < 6 ORDER BY rental.return_date LIMIT 4"
```

```python
## Using Left join command to create table with country_id,last_update, city_id and cities that are in Algeria and ordered alphabetically by the city name.
```

```python
!psql -c "SELECT country.country_id, country.country,country.last_update, city.city,city.city_id  \
            FROM pagila.country LEFT JOIN pagila.city ON country.country_id = city.country_id \
              WHERE country.country = 'Algeria' ORDER BY city LIMIT 10"
```

```python
## Using RIGHT JOIN command to get table of actors ordered by their first name and paired with the film id's they were in.
```

```python
!psql -c "SELECT actor.actor_id, actor.first_name,actor.last_name, film_actor.film_id \
            FROM pagila.actor RIGHT JOIN pagila.film_actor ON actor.actor_id = film_actor.actor_id \
            WHERE actor.actor_id < 20 ORDER BY actor.first_name LIMIT 30"
            
```

```python
## Usage of FULL OUter join to retrieve  matches for country_id, country, last-update, city, and city_id for cities that aren't Algeria and are ordered by city name alphabetically.
```

```python
!psql -c "SELECT country.country_id, country.country,country.last_update, city.city,city.city_id  \
            FROM pagila.country FULL OUTER JOIN pagila.city ON country.country_id = city.country_id \
              WHERE country.country != 'Algeria' ORDER BY city LIMIT 10"
```

```python
## This self-join matches addresses to the same district.
```

```python
!psql -c "SELECT A.address as address1, B.address as address2, A.district \
           FROM pagila.address A, pagila.address B \
           WHERE A.address_id <> B.address_id \
           AND A.district = B.district \
           ORDER BY A.district LIMIT 10 "
```
